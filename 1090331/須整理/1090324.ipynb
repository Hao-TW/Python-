```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
from random import randint
from time import time
```

```python
url = 'http://www.imdb.com/search/title?release_date=2017&sort=num_votes,desc&page=1'

response = get(url)
print(response.text[:300])
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')    #html_soup 為解析
type(html_soup)
movie_containers = html_soup.find_all('div',class_ = 'lister-item mode-advanced')


print(type(movie_containers))
print(len(movie_containers))
```

```python
first_movie = movie_containers[0]  #movie message
print(first_movie.div)
print("電影名稱: ",first_movie.h3.a.text)
```

```python
first_year = first_movie.h3.find('span',class_ = 'lister-item-year text-muted unbold')
first_year = first_year.text[1:5]  #[]內為切片
print("發行年份: ", first_year)
```

```python
movie_time = first_movie.p
movie_time = movie_time.text
print("電影時間:", movie_time)
```

```python
first_imdb = float(first_movie.strong.text)

print("IMDB評分:",first_imdb)
print(type(first_imdb))
```

```python
first_mscore = first_movie.find('span',class_ = 'metascore favorable')
first_mscore = int(first_mscore.text)
print('Metascore 評分 :',first_mscore)
```

```python
first_votes = first_movie.find('span',attrs = {'name':'nv'})
print (first_votes.text)
print ("投票數:" +first_votes['data-value'])
```

```python
names=[]
years=[]
imdb_ratings=[]
metascore=[]
votes=[]
for container in movie_containers:
    if container.find('div',class_='ratings-metascore') is not None:
        # The name
        name = container.h3.a.text
        names.append(name)
        # The year
        year = container.h3.find('span',class_ = 'lister-item-year').text
        years.append(year)
        # The IMDB ratings
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
        # The number of votes
        vote = container.find('span', attrs ={'name':'nv'})['data-value']
        votes.append(int(vote))
        
test_df = pd.DataFrame({'movie': names,
                       'year': years,
                       'imdb': imdb_ratings,
                        'metascore':metascore,
                       'votes': votes})
print(test_df.info())

test_df['year'].unique()
test_df.loc[:, 'year'] = test_df['year'].str[-5:-1].astype(int)
test_df.to_csv('movie_ratings.csv')
```
