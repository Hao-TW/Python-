```python
import numpy as np
from matplotlib import pyplot as plt


sale = np.array([22, 22, 17, 21, 22, 19, 21, 19, 20, 15, 24, 19])
profit = np.array([12, 15, 11, 17, 11, 12, 11, 10, 16, 16, 13, 11])
plt.xlim([0,13])
plt.ylim([0,25])

newa_ticks = np.linspace(0, 12, 13)
plt.xticks(new_ticks)
newb_ticks = np.linspace(0, 24, 13)
plt.yticks(newb_ticks)



plt.xlabel('Month', size=14)
plt.ylabel('Millin Dollars', size=14)
plt.title('sales and profit in one year', size=20)

plt.plot(sale,c='b',marker='o')
plt.plot(profit,c='r',marker='o')
plt.show()
```
