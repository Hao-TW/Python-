```python
# Assertion 斷言，程式進行到某個時間點，斷定其必然是某種狀態，具體而言就是斷定這個時間點上python要進行斷言測試，可以使用assert

class Account:
    def __init__(self,number,name):
        self.number = number
        self.name = name
        self.balance = 0
    def deposit(self,amount):
        assert amount > 0 ,"必須大於0的正數"
        self.balance += amount
    def withdraw(self,amount):
        assert amount > 0, "必須大於0的正數"
        if amount <= self.balance:
            self.balance -= amount
        else:
            raise RuntimeError("balance not enough")

a = Account('E122','Peter')
a.deposit(-1) #Assertion: 必須大於0

    
```

```python
from selenium import webdriver
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import time

#開啟 browser
def openChrome():
    option =  webdriver.Chrome("/Users/ASUS/chromedriver")
    return option

#授權操作
def operationAuth(driver):
    url = 'https://www.baidu.com/'
    driver.get(url)

    #找到輸入區輸入查詢內容
    #elem = driver.find_element_by_id('kw')
    elem = driver.find_element_by_name('wd')
    elem.send_keys('selenium')
    
    #提交表單
    driver.find_element_by_xpath("//*[@id='su']").click()
    print('查詢操做完畢')
    
if __name__ == '__main__':
    driver = openChrome()
    operationAuth(driver)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import pyautogui
from time import sleep

#滑鼠操作
# context_click(elem) 右鍵
# double_click(elem) 雙擊原宿，地圖WEB可實現放大功能
# drag_and_drop(source,target)移動滑鼠按下左鍵移動到目標
# move_to_element(elem)滑鼠移動到一個元素上
# click_and_hold(elem)按下左鍵在任一元素上

driver = webdriver.Chrome('/Users/ASUS/chromedriver')
driver.get('https://jianshu.com/')

#選擇元素
wait = WebDriverWait(driver,10)

#右鍵單擊圖片
img = wait.until(EC.element_to_be_clickable((By.TAG_NAME,'img')))

#執行滑鼠動作
actions = ActionChains(driver)

#找到圖片後右鍵
actions.context_click(img)
actions.perform()

#發送鍵盤按鍵，另存圖片
pyautogui.typewrite(['down','down','down','down','down','down','enter','enter'])

sleep(1)
pyautogui.typewrite(['enter'])
```
